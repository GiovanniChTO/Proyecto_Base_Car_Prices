```python
# Normalizar nombres de columnas de texto
text_cols = ['make', 'model', 'trim', 'body', 'transmission', 'color', 'interior', 'seller']

for col in text_cols:
    df[col] = df[col].astype(str).str.lower().str.strip()
    df[col] = df[col].replace('nan', np.nan)
```

```python
# Ejemplo: corregir 'automatico' → 'automatic'
df['transmission'] = df['transmission'].replace('automatico', 'automatic')

# Corregir 'sedan' mal escrito
df['body'] = df['body'].replace('sedann', 'sedan')

# Eliminar filas con valores claramente mal ubicados (ej. año = 0)
df = df[df['year'] > 1900]
```

```python
df['year'] = df['year'].astype(int)
df['condition'] = pd.to_numeric(df['condition'], errors='coerce')
df['odometer'] = pd.to_numeric(df['odometer'], errors='coerce')
df['mmr'] = pd.to_numeric(df['mmr'], errors='coerce')
df['sellingprice'] = pd.to_numeric(df['sellingprice'], errors='coerce')
```

```python
df = df[(df['sellingprice'] >= 500) & (df['sellingprice'] <= 100000)]
df = df[(df['year'] >= 1980) & (df['year'] <= 2024)]
```

```python
def validar_columna(df, columna, valores_validos=None, rango=None):
    print(f"\n=== Validando: {columna} ===")
    if valores_validos:
        invalidos = df[~df[columna].isin(valores_validos)][columna].unique()
        print("Valores no esperados:", invalidos)
    if rango:
        fuera_rango = df[(df[columna] < rango[0]) | (df[columna] > rango[1])][columna].unique()
        print("Valores fuera de rango:", fuera_rango)
```
